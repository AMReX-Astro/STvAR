```python
from STvAR_v3 import *
from STvARMetricObjects import *
from STvARFiniteDifference import *
from STvARAdvection import *
from STvARSymbolParsing import *
from STvARIndexing import*
```

```python
DiffOrder = 4
upDiffOrder = DiffOrder
dnDiffOrder = DiffOrder
DiffOrderKO = DiffOrder + 2
KOsigmaVal = 1.0
dim = 3
```

```python
gridvars = []
vartypes = []

chi = stvar('chi',  gridvar = True, vartype = 'state_fab')
gamtilde_LL = stvarrank2('gamtilde_LL',  gridvar = True, vartype = 'state_fab', sym = 'sym01')
Khat = stvar('Khat', gridvar = True, vartype = 'state_fab')
Atilde_LL = stvarrank2('Atilde_LL', gridvar = True, vartype = 'state_fab', sym = 'sym01')
theta = stvar('theta', gridvar = True, vartype = 'state_fab')
Gamhat_U = stvarrank1('Gamhat_U', gridvar = True, vartype ='state_fab')

alpha = stvar('alpha', gridvar = True, vartype = 'state_fab')
beta_U = stvarrank1('beta_U', gridvar = True, vartype = 'state_fab')
B_U = stvarrank1('B_U', gridvar = True, vartype = 'state_fab')
```

# RHS

```python
RHSString = ""
RHSString += chi.symb2isymb()
RHSString += gamtilde_LL.symb2isymb()
RHSString += Khat.symb2isymb()
RHSString += Atilde_LL.symb2isymb()
RHSString += theta.symb2isymb()
RHSString += Gamhat_U.symb2isymb()

RHSString += alpha.symb2isymb()
RHSString += beta_U.symb2isymb()
RHSString += B_U.symb2isymb()
```

```python
dDchi_L = chi.diff('dD', Accuracy = DiffOrder)
dupDchi_L = chi.diff('dup1D', Accuracy = upDiffOrder)
ddnDchi_L = chi.diff('ddn1D', Accuracy = dnDiffOrder)
dDDchi_LL = chi.diff('dDD', Accuracy = DiffOrder)
dKODFullchi = chi.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDchi = AdvectiveD(chi, beta_U)
```

```python
RHSString += dDchi_L.symb2isymb()
RHSString += dupDchi_L.symb2isymb()
RHSString += ddnDchi_L.symb2isymb()
RHSString += dDDchi_LL.symb2isymb()
RHSString += dKODFullchi.symb2isymb()
RHSString += AdvDchi.symb2expr()
```

```python
dDgamtilde_LLL = gamtilde_LL.diff('dD', Accuracy = DiffOrder)
dupDgamtilde_LLL = gamtilde_LL.diff('dup1D', Accuracy = upDiffOrder)
ddnDgamtilde_LLL = gamtilde_LL.diff('ddn1D', Accuracy = dnDiffOrder)
dDDgamtilde_LLLL = gamtilde_LL.diff('dDD', Accuracy = DiffOrder)
dKODFullgamtilde_LL = gamtilde_LL.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDgamtilde_LL = AdvectiveD(gamtilde_LL, beta_U)
```

```python
RHSString += dDgamtilde_LLL.symb2isymb()
RHSString += dupDgamtilde_LLL.symb2isymb()
RHSString += ddnDgamtilde_LLL.symb2isymb()
RHSString += dDDgamtilde_LLLL.symb2isymb()
RHSString += dKODFullgamtilde_LL.symb2isymb()
RHSString += AdvDgamtilde_LL.symb2expr()
```

```python
dDKhat_L = Khat.diff('dD', Accuracy = DiffOrder)
dupDKhat_L = Khat.diff('dup1D', Accuracy = upDiffOrder)
ddnDKhat_L = Khat.diff('ddn1D', Accuracy = dnDiffOrder)
dDDKhat_LL = Khat.diff('dDD', Accuracy = DiffOrder)
dKODFullKhat = Khat.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDKhat = AdvectiveD(Khat, beta_U)
```

```python
RHSString += dDKhat_L.symb2isymb()
RHSString += dupDKhat_L.symb2isymb()
RHSString += ddnDKhat_L.symb2isymb()
RHSString += dDDKhat_LL.symb2isymb()
RHSString += dKODFullKhat.symb2isymb()
RHSString += AdvDKhat.symb2expr()
```

```python
dDAtilde_LLL = Atilde_LL.diff('dD', Accuracy = DiffOrder)
dupDAtilde_LLL = Atilde_LL.diff('dup1D', Accuracy = upDiffOrder)
ddnDAtilde_LLL = Atilde_LL.diff('ddn1D', Accuracy = dnDiffOrder)
dDDAtilde_LLLL = Atilde_LL.diff('dDD', Accuracy = DiffOrder)
dKODFullAtilde_LL = Atilde_LL.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDAtilde_LL = AdvectiveD(Atilde_LL, beta_U)
```

```python
RHSString += dDAtilde_LLL.symb2isymb()
RHSString += dupDAtilde_LLL.symb2isymb()
RHSString += ddnDAtilde_LLL.symb2isymb()
RHSString += dDDAtilde_LLLL.symb2isymb()
RHSString += dKODFullAtilde_LL.symb2isymb()
RHSString += AdvDAtilde_LL.symb2expr()
```

```python
dDtheta_L = theta.diff('dD', Accuracy = DiffOrder)
dupDtheta_L = theta.diff('dup1D', Accuracy = upDiffOrder)
ddnDtheta_L = theta.diff('ddn1D', Accuracy = dnDiffOrder)
dDDtheta_LL = theta.diff('dDD', Accuracy = DiffOrder)
dKODFulltheta = theta.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDtheta = AdvectiveD(theta, beta_U)
```

```python
RHSString += dDtheta_L.symb2isymb()
RHSString += dupDtheta_L.symb2isymb()
RHSString += ddnDtheta_L.symb2isymb()
RHSString += dDDtheta_LL.symb2isymb()
RHSString += dKODFulltheta.symb2isymb()
RHSString += AdvDtheta.symb2expr()
```

```python
dDGamhat_UL = Gamhat_U.diff('dD', Accuracy = DiffOrder)
dupDGamhat_UL = Gamhat_U.diff('dup1D', Accuracy = upDiffOrder)
ddnDGamhat_UL = Gamhat_U.diff('ddn1D', Accuracy = dnDiffOrder)
dDDGamhat_ULL = Gamhat_U.diff('dDD', Accuracy = DiffOrder)
dKODFullGamhat_U = Gamhat_U.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDGamhat_U = AdvectiveD(Gamhat_U, beta_U)
```

```python
RHSString += dDGamhat_UL.symb2isymb()
RHSString += dupDGamhat_UL.symb2isymb()
RHSString += ddnDGamhat_UL.symb2isymb()
RHSString += dDDGamhat_ULL.symb2isymb()
RHSString += dKODFullGamhat_U.symb2isymb()
RHSString += AdvDGamhat_U.symb2expr()
```

```python
dDalpha_L = alpha.diff('dD', Accuracy = DiffOrder)
dupDalpha_L = alpha.diff('dup1D', Accuracy = upDiffOrder)
ddnDalpha_L = alpha.diff('ddn1D', Accuracy = dnDiffOrder)
dDDalpha_LL = alpha.diff('dDD', Accuracy = DiffOrder)
dKODFullalpha = alpha.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDalpha = AdvectiveD(alpha, beta_U)
```

```python
RHSString += dDalpha_L.symb2isymb()
RHSString += dupDalpha_L.symb2isymb()
RHSString += ddnDalpha_L.symb2isymb()
RHSString += dDDalpha_LL.symb2isymb()
RHSString += dKODFullalpha.symb2isymb()
RHSString += AdvDalpha.symb2expr()
```

```python
dDbeta_UL = beta_U.diff('dD', Accuracy = DiffOrder)
dupDbeta_UL = beta_U.diff('dup1D', Accuracy = upDiffOrder)
ddnDbeta_UL = beta_U.diff('ddn1D', Accuracy = dnDiffOrder)
dDDbeta_ULL = beta_U.diff('dDD', Accuracy = DiffOrder)
dKODFullbeta_U = beta_U.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDbeta_U = AdvectiveD(beta_U, beta_U)
```

```python
RHSString += dDbeta_UL.symb2isymb()
RHSString += dupDbeta_UL.symb2isymb()
RHSString += ddnDbeta_UL.symb2isymb()
RHSString += dDDbeta_ULL.symb2isymb()
RHSString += dKODFullbeta_U.symb2isymb()
RHSString += AdvDbeta_U.symb2expr()
```

```python
dDB_UL = B_U.diff('dD', Accuracy = DiffOrder)
dupDB_UL = B_U.diff('dup1D', Accuracy = upDiffOrder)
ddnDB_UL = B_U.diff('ddn1D', Accuracy = dnDiffOrder)
dDDB_ULL = B_U.diff('dDD', Accuracy = DiffOrder)
dKODFullB_U = B_U.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDB_U = AdvectiveD(B_U, beta_U)
```

```python
RHSString += dDB_UL.symb2isymb()
RHSString += dupDB_UL.symb2isymb()
RHSString += ddnDB_UL.symb2isymb()
RHSString += dDDB_ULL.symb2isymb()
RHSString += dKODFullB_U.symb2isymb()
RHSString += AdvDB_U.symb2expr()
```

```python
Gamtilde_LLL = stvarrank3('Gamtilde_LLL', sym = 'sym12')

for i in range(3):
    for j in range(3):
        for k in range(3):
            Gamtilde_LLL.expr[i][j][k] += sp.Rational(1,2)*(dDgamtilde_LLL.symb[i][k][j] + dDgamtilde_LLL.symb[i][j][k] - dDgamtilde_LLL.symb[j][k][i])
```

```python
Gamtilde_ULL = stvarrank3('Gamtilde_ULL', sym = 'sym12')

for i in range(3):
    for j in range(3):
        for k in range(3):
            for l in range(3):
                Gamtilde_ULL.expr[k][i][j] += gamtilde_UU.symb[k][l]*Gamtilde_LLL.symb[l][i][j]
```

```python
RHSString += Gamtilde_LLL.symb2expr()
RHSString += Gamtilde_ULL.symb2expr()
```

```python
Atilde_UU = stvarrank2("Atilde_UU", sym = 'sym01')
for i in range(3):
    for j in range(3):
        for k in range(3):
            for l in range(3):
                Atilde_UU.expr[i][j] += gamtilde_UU.symb[i][k]*gamtilde_UU.symb[j][l]*Atilde_LL.symb[k][l]
                
Atilde_UL = stvarrank2("Atilde_UL")
for i in range(3):
    for j in range(3):
        for k in range(3):
            Atilde_UL.expr[i][j] += gamtilde_UU.symb[i][k]*Atilde_LL.symb[k][j]
```

```python
RHSString += Atilde_UU.symb2expr()
RHSString += Atilde_UL.symb2expr()
```

```python
Gamtilde_U = stvarrank1('Gamtilde_U')
for i in range(3):
    for j in range(3):
        for k in range(3):
            Gamtilde_U.expr[i] += gamtilde_UU.symb[j][k]*Gamtilde_ULL.symb[i][j][k]
```

```python
RHSString += Gamtilde_U.symb2expr()
```

```python
Ztilde_U = stvarrank1('Ztilde_U')
for i in range(3):
    Ztilde_U.expr[i] += sp.Rational(1,2)*(Gamhat_U.symb[i] - Gamtilde_U.symb[i])
```

```python
RHSString += Ztilde_U.symb2expr()
```

```python
s = stvar('s')
s.expr = 1

c = stvar('c')
c.expr = 0

kappa = stvar('kappa')
kappa.expr = 0
```

```python
RHSString += s.symb2expr()
RHSString += c.symb2expr()
RHSString += kappa.symb2expr()
```

```python
Gamhatc_U = stvarrank1('Gamhatc_U')
for i in range(3):
    Gamhatc_U.expr[i] += c.symb*Gamhat_U.symb[i] + (1-c.symb)*Gamtilde_U.symb[i]
```

```python
RHSString += Gamhatc_U.symb2expr()
```

```python
Gamhatsc_U = stvarrank1('Gamhatsc_U')
for i in range(3):
    Gamhatsc_U.expr[i] += s.symb*c.symb*Gamhat_U.symb[i] + (1-s.symb*c.symb)*Gamtilde_U.symb[i]
```

```python
RHSString += Gamhatsc_U.symb2expr()
```

```python
Rhat_LL = stvarrank2('Rhat_LL', sym = 'sym01')

for i in range(3):
    for j in range(3):
        for k in range(3):
            for l in range(3):
                Rhat_LL.expr[i][j] += -sp.Rational(1,2)*chi.symb*gamtilde_UU.symb[k][l]*dDDgamtilde_LLLL.symb[i][j][k][l]

for i in range(3):
    for j in range(3):
        for k in range(3):
            Rhat_LL.expr[i][j] += sp.Rational(1,2)*chi.symb*(gamtilde_LL.symb[k][i]*dDGamhat_UL.symb[k][j]+gamtilde_LL.symb[k][j]*dDGamhat_UL.symb[k][i])

            
for i in range(3):
    for j in range(3):
        for k in range(3):
            Rhat_LL.expr[i][j] += sp.Rational(1,2)*chi.symb*(Gamhatc_U.symb[k]*Gamtilde_LLL.symb[i][j][k]+Gamhatc_U.symb[k]*Gamtilde_LLL.symb[j][i][k])

for i in range(3):
    for j in range(3):
        for k in range(3):
            for l in range(3):
                for m in range(3):
                    Rhat_LL.expr[i][j] += chi.symb*gamtilde_UU.symb[k][l]*Gamtilde_ULL.symb[m][k][i]*Gamtilde_LLL.symb[m][l][j]
                    Rhat_LL.expr[i][j] += chi.symb*gamtilde_UU.symb[k][l]*(Gamtilde_ULL.symb[m][k][i]*Gamtilde_LLL.symb[j][m][l]+Gamtilde_ULL.symb[m][k][j]*Gamtilde_LLL.symb[i][m][l])
                   
                    
for i in range(3):
    for j in range(3):
        Rhat_LL.expr[i][j] += sp.Rational(1,2)*(dDDchi_LL.symb[i][j]-(sp.Rational(1,2)/chi.symb)*dDchi_L.symb[i]*dDchi_L.symb[j])
        
for i in range(3):
    for j in range(3):
        for k in range(3):
            for l in range(3):
                Rhat_LL.expr[i][j] += sp.Rational(1,2)*gamtilde_LL.symb[i][j]*gamtilde_UU.symb[k][l]*(dDDchi_LL.symb[l][k]-(sp.Rational(3,2)/chi.symb)*dDchi_L.symb[k]*dDchi_L.symb[l])
                
for i in range(3):
    for j in range(3):
        for k in range(3):
            Rhat_LL.expr[i][j] += -sp.Rational(1,2)*(Gamtilde_ULL.symb[k][i][j]*dDchi_L.symb[k]+gamtilde_LL.symb[i][j]*Gamhatc_U.symb[k]*dDchi_L.symb[k])
```

```python
RHS_theta = stvar("theta", gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
RHS_theta.expr += AdvDtheta.symb     
RHS_theta.expr += KOsigma.symb*dKODFulltheta.symb

RHS_theta.expr += alpha.symb*(sp.Rational(1,3)*Khat.symb**2 - (1 - sp.Rational(7,3)*s.symb)*Khat.symb*theta.symb - kappa.symb*theta.symb + sp.Rational(4,3)*s.symb*(theta.symb**2))

for i in range(3):
    for j in range(3):
        RHS_theta.expr += -sp.Rational(1,2)*alpha.symb*Atilde_LL.symb[i][j]*Atilde_UU.symb[i][j]
        RHS_theta.expr += sp.Rational(1,2)*alpha.symb*gamtilde_UU.symb[i][j]*Rhat_LL.symb[i][j]
        
for i in range(3):
    RHS_theta.expr += c.symb*Ztilde_U.symb[i]*(alpha.symb*dDchi_L.symb[i]-chi.symb*dDalpha_L.symb[i])
```

```python
RHSString += RHS_theta.setisymb()
```

```python
RHS_chi = stvar("chi", gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
RHS_chi.expr += AdvDchi.symb     
RHS_chi.expr += KOsigma.symb*dKODFullchi.symb

RHS_chi.expr += sp.Rational(2,3)*chi.symb*alpha.symb*(Khat.symb + 2*s.symb*theta.symb)
for i in range(3):
    RHS_chi.expr += -sp.Rational(2,3)*chi.symb*dDbeta_UL.symb[i][i]
```

```python
RHSString += RHS_chi.setisymb()
```

```python
RHS_Khat = stvar("Khat", gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
RHS_Khat.expr += AdvDKhat.symb     
RHS_Khat.expr += KOsigma.symb*dKODFullKhat.symb

RHS_Khat.expr += alpha.symb*((1-sp.Rational(2,3)*s.symb)*Khat.symb**2 - 2*(1-sp.Rational(5,3)*s.symb)*Khat.symb*theta.symb-(3-4*s.symb)*kappa.symb*theta.symb+sp.Rational(4,3)*s.symb*theta.symb**2)

for i in range(3):
    for j in range(3):
        RHS_Khat.expr += alpha.symb*s.symb*Atilde_LL.symb[i][j]*Atilde_UU.symb[i][j]
        
for i in range(3):
    RHS_Khat.expr += alpha.symb*2*(1-s.symb)*c.symb*Ztilde_U.symb[i]*dDchi_L.symb[i]
    
for i in range(3):
    RHS_Khat.expr += chi.symb*Gamhatsc_U.symb[i]*dDalpha_L.symb[i]
    
for i in range(3):
    for j in range(3):
        RHS_Khat.expr += gamtilde_UU.symb[i][j]*(-chi.symb*dDDalpha_LL.symb[i][j] + sp.Rational(1,2)*dDchi_L.symb[i]*dDalpha_L.symb[j])
        RHS_Khat.expr += gamtilde_UU.symb[i][j]*alpha.symb*((1-s.symb)*Rhat_LL.symb[i][j])
```

```python
RHSString += RHS_Khat.setisymb()
```

```python
RHS_Gamhat_U = stvarrank1('Gamhat_U', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")

for i in range(3):
    RHS_Gamhat_U.expr[i] += AdvDGamhat_U.symb[i]
    RHS_Gamhat_U.expr[i] += KOsigma.symb*dKODFullGamhat_U.symb[i]
    
for i in range(3):
    for j in range(3):
        for k in range(3):
            RHS_Gamhat_U.expr[i] += 2*alpha.symb*Gamtilde_ULL.symb[i][j][k]*Atilde_UU.symb[j][k]
            
for i in range(3):
    for j in range(3):
        RHS_Gamhat_U.expr[i] += -3*alpha.symb*Atilde_UU.symb[i][j]*dDchi_L.symb[j]/chi.symb
        RHS_Gamhat_U.expr[i] += 2*alpha.symb*gamtilde_UU.symb[i][j]*((1-sp.Rational(4,3))*dDtheta_L.symb[j]-sp.Rational(2,3)*dDKhat_L.symb[j])
        
for i in range(3):
    RHS_Gamhat_U.expr[i] += -2*alpha.symb*c.symb*(sp.Rational(2,3)*Khat.symb+sp.Rational(4,3)*s.symb*theta.symb+kappa.symb)*Ztilde_U.symb[i]
    
for i in range(3):
    for j in range(3):
        RHS_Gamhat_U.expr[i] += -2*c.symb*theta.symb*gamtilde_UU.symb[i][j]*dDalpha_L.symb[j]-2*Atilde_UU.symb[i][j]*dDalpha_L.symb[j]
        
for i in range(3):
    for j in range(3):
        for k in range(3):
            RHS_Gamhat_U.expr[i] += gamtilde_UU.symb[j][k]*dDDbeta_ULL.symb[i][j][k] + sp.Rational(1,3)*gamtilde_UU.symb[i][j]*dDDbeta_ULL.symb[k][j][k]
            
for i in range(3):
    for j in range(3):
        RHS_Gamhat_U.expr[i] += -Gamhatc_U.symb[j]*dDbeta_UL.symb[i][j]+sp.Rational(2,3)*Gamhatc_U.symb[i]*dDbeta_UL.symb[j][j]
    
```

```python
RHS_gamtilde_LL = stvarrank2("gamtilde_LL", gridvar = True, addtolist = False, sym = 'sym01', vartype = "state_fab", varprefix = "rhs_fab")

for i in range(3):
    for j in range(3):
        RHS_gamtilde_LL.expr[i][j] += AdvDgamtilde_LL.symb[i][j]
        RHS_gamtilde_LL.expr[i][j] += KOsigma.symb*dKODFullgamtilde_LL.symb[i][j]
        
for i in range(3):
    for j in range(3):
        RHS_gamtilde_LL.expr[i][j] += -2*alpha.symb*Atilde_LL.symb[i][j]
        
for i in range(3):
    for j in range(3):
        for k in range(3):
            RHS_gamtilde_LL.expr[i][j] += gamtilde_LL.symb[i][k]*dDbeta_UL.symb[k][j] + gamtilde_LL.symb[j][k]*dDbeta_UL.symb[k][i] - sp.Rational(2,3)*gamtilde_LL.symb[i][j]*dDbeta_UL.symb[k][k]
        
        
```

```python
RHSPartAtilde_LL = stvarrank2('RHSPartAtilde_LL', sym = 'sym01')

for i in range(3):
    for j in range(3):
        RHSPartAtilde_LL.expr[i][j] += -chi.symb*dDDalpha_LL.symb[i][j]
        
for i in range(3):
    for j in range(3):
        for k in range(3):
            RHSPartAtilde_LL.expr[i][j] += chi.symb*Gamtilde_ULL.symb[k][i][j]*dDalpha_L.symb[k]
            
for i in range(3):
    for j in range(3):
        RHSPartAtilde_LL.expr[i][j] += -sp.Rational(1,2)*(dDchi_L.symb[i]*dDalpha_L.symb[j]+dDchi_L.symb[j]*dDalpha_L.symb[i])
        
for i in range(3):
    for j in range(3):
        for k in range(3):
            for l in range(3):
                RHSPartAtilde_LL.expr[i][j] += -sp.Rational(1,2)*gamtilde_LL.symb[i][j]*gamtilde_UU.symb[k][l]*dDchi_L.symb[k]*dDalpha_L.symb[l]
            
for i in range(3):
    for j in range(3):
        for k in range(3):
            RHSPartAtilde_LL.expr[i][j] += c.symb*alpha.symb*Ztilde_U.symb[k]*(gamtilde_LL.symb[k][i]*dDchi_L.symb[j]+gamtilde_LL.symb[k][j]*dDchi_L.symb[i])

for i in range(3):
    for j in range(3):
        RHSPartAtilde_LL.expr[i][j] += alpha.symb*Rhat_LL.symb[i][j]
```
