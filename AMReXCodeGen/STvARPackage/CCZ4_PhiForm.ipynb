```python
RHS_theta = stvar("theta", gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")

RHS_theta.expr += AdvDtheta.symb     
RHS_theta.expr += KOsigma.symb*dKODFulltheta.symb

RHS_theta.expr += sp.Rational(1,2)*alpha.symb*(Rsclr.symb + 2*divZ.symb +sp.Rational(2,3)*(K.symb**2)-2*theta.symb*K.symb)

for i in range(3):
    for j in range(3):
        RHS_theta.expr += -sp.Rational(1,2)*alpha.symb*Atilde_LL.symb[i][j]*Atilde_UU.symb[i][j]
        
for i in range(3):
    RHS_theta.expr += -Z_U.symb[i]*dDalpha_L.symb[i]
    
RHS_theta.expr += -alpha.symb*kappa1.symb*(2.0 + kappa2.symb)*theta.symb
```

```python
RHSString += RHS_theta.setisymb()
```

```python
RHS_Gamhat_U = stvarrank1('Gamhat_U', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")

for i in range(3):
    RHS_Gamhat_U.expr[i] += AdvDGamhat_U.symb[i]
    RHS_Gamhat_U.expr[i] += KOsigma.symb*dKODFullGamhat_U.symb[i]
    
for i in range(3):
    for j in range(3):
        for k in range(3):
            RHS_Gamhat_U.expr[i] += 2*alpha.symb*Gamtilde_ULL.symb[i][j][k]*Atilde_UU.symb[j][k]    
            
for i in range(3):
    for j in range(3):
        RHS_Gamhat_U.expr[i] += 2*alpha.symb*((-3.0/phi.symb)*Atilde_UU.symb[i][j]*dDphi_L.symb[j]-sp.Rational(2,3)*gamtilde_UU.symb[i][j]*dDK_L.symb[j])
        
for i in range(3):
    for k in range(3):
        RHS_Gamhat_U.expr[i] += 2*gamtilde_UU.symb[k][i]*(alpha.symb*dDtheta_L.symb[k] - theta.symb*dDalpha_L.symb[k] - sp.Rational(2,3)*alpha.symb*K.symb*Z_L.symb[k])
        
for i in range(3):
    for j in range(3):
        RHS_Gamhat_U.expr[i] += -2*Atilde_UU.symb[i][j]*dDalpha_L.symb[j]
        
for i in range(3):
    for k in range(3):
        for l in range(3):
            RHS_Gamhat_U.expr[i] += gamtilde_UU.symb[k][l]*dDDbeta_ULL.symb[i][k][l] + sp.Rational(1,3)*gamtilde_UU.symb[i][k]*dDDbeta_ULL.symb[l][k][l]
            
for i in range(3):
    for k in range(3):
        RHS_Gamhat_U.expr[i] += sp.Rational(2,3)*Gamtilde_U.symb[i]*dDbeta_UL.symb[k][k] - Gamtilde_U.symb[i]*dDbeta_UL.symb[i][k]
        
    
for i in range(3):
    for j in range(3):
        for k in range(3):
            RHS_Gamhat_U.expr[i] += 2*kappa3.symb*(sp.Rational(2,3)*gamtilde_UU.symb[i][j]*Z_L.symb[j]*dDbeta_UL.symb[k][k]-gamtilde_UU.symb[j][k]*Z_L.symb[j]*dDbeta_UL.symb[i][k])
            
for i in range(3):
    for j in range(3):
        RHS_Gamhat_U.expr[i] += -2*alpha.symb*kappa1.symb*gamtilde_UU.symb[i][j]*Z_L.symb[j]
```
