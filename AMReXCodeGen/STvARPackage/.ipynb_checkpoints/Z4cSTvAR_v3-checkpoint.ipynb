```python
from STvAR_v3 import *
from STvARMetricObjects import *
from STvARFiniteDifference import *
from STvARAdvection import *
from STvARSymbolParsing import *
from STvARIndexing import*
```

```python
DiffOrder = 4
DiffOrderKO = DiffOrder + 2
KOsigma = 0.3
dim = 3
```

```python
gridvars = []
vartypes = []

chi = stvar('chi',  gridvar = True, vartype = 'state_fab')
gamtilde_LL = stvarrank2('gamtilde_LL',  gridvar = True, vartype = 'state_fab', sym = 'sym01')
Khat = stvar('Khat', gridvar = True, vartype = 'state_fab')
Atilde_LL = stvarrank2('Atilde_LL', gridvar = True, vartype = 'state_fab', sym = 'sym01')
theta = stvar('theta', gridvar = True, vartype = 'state_fab')
Gamtilde_U = stvarrank1('Gamtilde_U', gridvar = True, vartype ='state_fab')

alpha = stvar('alpha', gridvar = True, vartype = 'state_fab')
beta_U = stvarrank1('beta_U', gridvar = True, vartype = 'state_fab')
```

```python
declareStates = ""
declareStates += chi.symb2isymb()
declareStates += gamtilde_LL.symb2isymb()
declareStates += Khat.symb2isymb()
declareStates += Atilde_LL.symb2isymb()
declareStates += theta.symb2isymb()
declareStates += Gamtilde_U.symb2isymb()

declareStates += alpha.symb2isymb()
declareStates += beta_U.symb2isymb()
```

```python
declareStates = declareStates.replace("state_fabIdx", "Idx")
```

```python
declareDerivatives = ""
```

```python
dDchi_L = chi.diff('dD', Accuracy = DiffOrder)
dupDchi_L = chi.diff('dup1D', Accuracy = DiffOrder)
ddnDchi_L = chi.diff('ddn1D', Accuracy = DiffOrder)
dDDchi_LL = chi.diff('dDD', Accuracy = DiffOrder)
dKODFullchi = chi.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDchi = AdvectiveD(chi, beta_U)
```

```python
declareDerivatives += dDchi_L.symb2isymb()
declareDerivatives += dupDchi_L.symb2isymb()
declareDerivatives += ddnDchi_L.symb2isymb()
declareDerivatives += dDDchi_LL.symb2isymb()
declareDerivatives += dKODFullchi.symb2isymb()
declareDerivatives += AdvDchi.symb2expr()
```

```python
dDgamtilde_LLL = gamtilde_LL.diff('dD', Accuracy = DiffOrder)
dupDgamtilde_LLL = gamtilde_LL.diff('dup1D', Accuracy = DiffOrder)
ddnDgamtilde_LLL = gamtilde_LL.diff('ddn1D', Accuracy = DiffOrder)
dDDgamtilde_LLLL = gamtilde_LL.diff('dDD', Accuracy = DiffOrder)
dKODFullgamtilde_LL = gamtilde_LL.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDgamtilde_LL = AdvectiveD(gamtilde_LL, beta_U)
```

```python
declareDerivatives += dDgamtilde_LLL.symb2isymb()
declareDerivatives += dupDgamtilde_LLL.symb2isymb()
declareDerivatives += ddnDgamtilde_LLL.symb2isymb()
declareDerivatives += dDDgamtilde_LLLL.symb2isymb()
declareDerivatives += dKODFullgamtilde_LL.symb2isymb()
declareDerivatives += AdvDgamtilde_LL.symb2expr()
```

```python
dDKhat_L = Khat.diff('dD', Accuracy = DiffOrder)
dupDKhat_L = Khat.diff('dup1D', Accuracy = DiffOrder)
ddnDKhat_L = Khat.diff('ddn1D', Accuracy = DiffOrder)
dDDKhat_LL = Khat.diff('dDD', Accuracy = DiffOrder)
dKODFullKhat = Khat.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDKhat = AdvectiveD(Khat, beta_U)
```

```python
declareDerivatives += dDKhat_L.symb2isymb()
declareDerivatives += dupDKhat_L.symb2isymb()
declareDerivatives += ddnDKhat_L.symb2isymb()
declareDerivatives += dDDKhat_LL.symb2isymb()
declareDerivatives += dKODFullKhat.symb2isymb()
declareDerivatives += AdvDKhat.symb2expr()
```

```python
dDAtilde_LLL = Atilde_LL.diff('dD', Accuracy = DiffOrder)
dupDAtilde_LLL = Atilde_LL.diff('dup1D', Accuracy = DiffOrder)
ddnDAtilde_LLL = Atilde_LL.diff('ddn1D', Accuracy = DiffOrder)
dDDAtilde_LLLL = Atilde_LL.diff('dDD', Accuracy = DiffOrder)
dKODFullAtilde_LL = Atilde_LL.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDAtilde_LL = AdvectiveD(Atilde_LL, beta_U)
```

```python
declareDerivatives += dDAtilde_LLL.symb2isymb()
declareDerivatives += dupDAtilde_LLL.symb2isymb()
declareDerivatives += ddnDAtilde_LLL.symb2isymb()
declareDerivatives += dDDAtilde_LLLL.symb2isymb()
declareDerivatives += dKODFullAtilde_LL.symb2isymb()
declareDerivatives += AdvDAtilde_LL.symb2expr()
```

```python
dDtheta_L = theta.diff('dD', Accuracy = DiffOrder)
dupDtheta_L = theta.diff('dup1D', Accuracy = DiffOrder)
ddnDtheta_L = theta.diff('ddn1D', Accuracy = DiffOrder)
dDDtheta_LL = theta.diff('dDD', Accuracy = DiffOrder)
dKODFulltheta = theta.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDtheta = AdvectiveD(theta, beta_U)
```

```python
declareDerivatives += dDtheta_L.symb2isymb()
declareDerivatives += dupDtheta_L.symb2isymb()
declareDerivatives += ddnDtheta_L.symb2isymb()
declareDerivatives += dDDtheta_LL.symb2isymb()
declareDerivatives += dKODFulltheta.symb2isymb()
declareDerivatives += AdvDtheta.symb2expr()
```

```python
dDGamtilde_UL = Gamtilde_U.diff('dD', Accuracy = DiffOrder)
dupDGamtilde_UL = Gamtilde_U.diff('dup1D', Accuracy = DiffOrder)
ddnDGamtilde_UL = Gamtilde_U.diff('ddn1D', Accuracy = DiffOrder)
dDDGamtilde_ULL = Gamtilde_U.diff('dDD', Accuracy = DiffOrder)
dKODFullGamtilde_U = Gamtilde_U.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDGamtilde_U = AdvectiveD(Gamtilde_U, beta_U)
```

```python
declareDerivatives += dDGamtilde_UL.symb2isymb()
declareDerivatives += dupDGamtilde_UL.symb2isymb()
declareDerivatives += ddnDGamtilde_UL.symb2isymb()
declareDerivatives += dDDGamtilde_ULL.symb2isymb()
declareDerivatives += dKODFullGamtilde_U.symb2isymb()
declareDerivatives += AdvDGamtilde_U.symb2expr()
```

```python
dDalpha_L = alpha.diff('dD', Accuracy = DiffOrder)
dupDalpha_L = alpha.diff('dup1D', Accuracy = DiffOrder)
ddnDalpha_L = alpha.diff('ddn1D', Accuracy = DiffOrder)
dDDalpha_LL = alpha.diff('dDD', Accuracy = DiffOrder)
dKODFullalpha = alpha.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDalpha = AdvectiveD(alpha, beta_U)
```

```python
declareDerivatives += dDalpha_L.symb2isymb()
declareDerivatives += dupDalpha_L.symb2isymb()
declareDerivatives += ddnDalpha_L.symb2isymb()
declareDerivatives += dDDalpha_LL.symb2isymb()
declareDerivatives += dKODFullalpha.symb2isymb()
declareDerivatives += AdvDalpha.symb2expr()
```

```python
dDbeta_UL = beta_U.diff('dD', Accuracy = DiffOrder)
dupDbeta_UL = beta_U.diff('dup1D', Accuracy = DiffOrder)
ddnDbeta_UL = beta_U.diff('ddn1D', Accuracy = DiffOrder)
dDDbeta_ULL = beta_U.diff('dDD', Accuracy = DiffOrder)
dKODFullbeta_U = beta_U.diff('dKODFull', Accuracy = DiffOrderKO)
AdvDbeta_U = AdvectiveD(beta_U, beta_U)
```

```python
declareDerivatives += dDbeta_UL.symb2isymb()
declareDerivatives += dupDbeta_UL.symb2isymb()
declareDerivatives += ddnDbeta_UL.symb2isymb()
declareDerivatives += dDDbeta_ULL.symb2isymb()
declareDerivatives += dKODFullbeta_U.symb2isymb()
declareDerivatives += AdvDbeta_U.symb2expr()
```

```python
declareDerivatives = declareDerivatives.replace("state_fabIdx", "Idx")
```

```python
metricObjects = ""
```

```python
gamtilde_UU = inversemetric(gamtilde_LL)
```

```python
Gamtilde_LLL = ChristoffelLLL(gamtilde_LL)
Gamtilde_ULL = ChristoffelULL(gamtilde_LL)
Gam_ULL = ConformalChristoffelULL(chi, gamtilde_LL)
```

```python
metricObjects += gamtilde_UU.symb2expr()
metricObjects += Gamtilde_LLL.symb2expr()
metricObjects += Gamtilde_ULL.symb2expr()
metricObjects += Gam_ULL.symb2expr()
```

```python
CovDDalpha_LL = CovariantD(alpha, Gam_ULL, dstring = 'dDD')
"""CovDDalpha_LL = stvarrank2('CovDDalpha_LL')
for i in range(3):
    for j in range(3):
        CovDDalpha_LL.expr[i][j] += dDDalpha_LL.symb[i][j]
        for k in range(3):
            CovDDalpha_LL.expr[i][j] += -Gam_ULL.symb[k][i][j]*dDalpha_L.symb[k]
"""
```

```python
metricObjects += CovDDalpha_LL.symb2expr()
```

```python
gam_LL = stvarrank2('gam_LL')
gam_LL.expr = (chi.symb**(-1))*gamtilde_LL.symb

gam_UU = stvarrank2('invgam_UU')
gam_UU.expr = chi.symb*gamtilde_UU.symb
```

```python
metricObjects += gam_LL.symb2expr()
metricObjects += gam_UU.symb2expr()
```

```python
Atilde_UU = stvarrank2("Atilde_UU")
for i in range(3):
    for j in range(3):
        for k in range(3):
            for l in range(3):
                Atilde_UU.expr[i][j] += gamtilde_UU.symb[i][k]*gamtilde_UU.symb[j][l]*Atilde_LL.symb[k][l]
                
Atilde_UL = stvarrank2("Atilde_UL")
for i in range(3):
    for j in range(3):
        for k in range(3):
            Atilde_UL.expr[i][j] += gamtilde_UU.symb[i][k]*Atilde_LL.symb[k][j]
```

```python
metricObjects += Atilde_UU.symb2expr()
metricObjects += Atilde_UL.symb2expr()
```

```python
CovDDalphaTF_LL = stvarrank2('CovDDalphaTF_LL')

for i in range(3):
    for j in range(3):
        CovDDalphaTF_LL.expr[i][j] += CovDDalpha_LL.symb[i][j]
        for k in range(3):
            for l in range(3):
                CovDDalphaTF_LL.expr[i][j] += -1/3*gam_LL.symb[i][j]*(gam_UU.symb[k][l]*CovDDalpha_LL.symb[k][l])
```

```python
metricObjects += CovDDalphaTF_LL.symb2expr()
```

```python
GamtildeD_U = ChristoffelDU(gamtilde_LL)
```

```python
metricObjects += GamtildeD_U.symb2expr()
```

```python
Rtilde_LL = Z4cRicciTildeTensorLL(gamtilde_LL, Gamtilde_U)
```

```python
metricObjects += Rtilde_LL.symb2expr()
```

```python
CovDtildechi_L = CovariantD(chi, Gamtilde_ULL, dstring = 'dD')
CovDtildeDtildechi_LL = CovariantD(chi, Gamtilde_ULL, dstring = 'dDD')
            
TrCovDtildeDtildechi = stvar('TrCovDtildeDtildechi')
for i in range(3):
    for j in range(3):
        TrCovDtildeDtildechi.expr += gamtilde_UU.symb[i][j]*CovDtildeDtildechi_LL.symb[i][j]
```

```python
metricObjects += CovDtildechi_L.symb2expr()
metricObjects += CovDtildeDtildechi_LL.symb2expr()
metricObjects += TrCovDtildeDtildechi.symb2expr()
```

```python
Rchi_LL = Z4cRicciTensorFromScalar(chi, gamtilde_LL, dstring = 'dD')       
"""
for i in range(3):
    for j in range(3):
        Rchi_LL.expr[i][j] += 1/2*(chi.symb**(-1))*(CovDDchi_LL.symb[i][j]+gamtilde_LL.symb[i][j]*TrCovDDchi.symb)
        Rchi_LL.expr[i][j] += -1/4*(chi.symb**(-2))*dDchi_L.symb[i]*dDchi_L.symb[j]
        for l in range(3):
            for m in range(3):
                Rchi_LL.expr[i][j] += -3/4*(chi.symb**(-2))*gamtilde_LL.symb[i][j]*gamtilde_UU.symb[l][m]*dDchi_L.symb[m]*dDchi_L.symb[l]
"""
```

```python
print(Rchi_LL.symb2expr())
```

```python
metricObjects += Rchi_LL.symb2expr()
```

```python
R_LL = stvarrank2('R_LL')
for i in range(3):
    for j in range(3):
        R_LL.expr[i][j] += Rtilde_LL.symb[i][j]+Rchi_LL.symb[i][j]
```

```python
metricObjects += R_LL.symb2expr()
```

```python
Rsclr = stvar('Rsclr')
for i in range(3):
    for j in range(3):
        Rsclr.expr += gam_UU.symb[i][j]*R_LL.symb[i][j] 
```

```python
metricObjects += Rsclr.symb2expr()
```

```python
RTF_LL = stvarrank2('RTF_LL')
for i in range(3):
    for j in range(3):
        RTF_LL.expr[i][j] = R_LL.symb[i][j]-1/3*gam_LL.symb[i][j]*Rsclr.symb
```

```python
metricObjects += RTF_LL.symb2expr()
```

```python
Dtildebeta_UL = CovariantD(beta_U, Gamtilde_ULL, 'dD')
```

```python
metricObjects += Dtildebeta_UL.symb2expr()
```

```python
divTildebeta = stvar('divTildebeta')
for i in range(3):
    divTildebeta.expr += Dtildebeta_UL.symb[i][i]
```

```python
metricObjects += divTildebeta.symb2expr()
```

```python
RHSObjects = ""
```

```python
RHS_chi = stvar("chi", gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
RHS_chi.expr = 2.0/3*chi.symb*(alpha.symb*(Khat.symb+2*theta.symb)-divTildebeta.symb)
RHS_chi.expr += AdvDchi.symb     
RHS_chi.expr += KOsigma*dKODFullchi.symb
```

```python
RHSObjects += RHS_chi.setisymb()
```

```python
RHS_gamtilde_LL = stvarrank2("gamtilde_LL", gridvar = True, addtolist = False, sym = 'sym01', vartype = "state_fab", varprefix = "rhs_fab")

for i in range(3):
    for j in range(3):
        RHS_gamtilde_LL.expr[i][j] += -2*alpha.symb*Atilde_LL.symb[i][j]
        RHS_gamtilde_LL.expr[i][j] += AdvDgamtilde_LL.symb[i][j]
        for k in range(3):
            RHS_gamtilde_LL.expr[i][j] += gamtilde_LL.symb[k][i]*dDbeta_UL.symb[k][j]+gamtilde_LL.symb[k][j]*dDbeta_UL.symb[k][i]
            RHS_gamtilde_LL.expr[i][j] += -2/3*gamtilde_LL.symb[i][j]*dDbeta_UL.symb[k][k]
            
for i in range(3):
    for j in range(3):
        RHS_gamtilde_LL.expr[i][j] += KOsigma*dKODFullgamtilde_LL.symb[i][j]
```

```python
RHSObjects += RHS_gamtilde_LL.setisymb()
```

```python
kappa1 = stvar('kappa1')
kappa2 = stvar('kappa2')

kappa1.expr = 0.02
kappa2.expr = 0

RHS_Khat = stvar('Khat', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
for i in range(3):
    for j in range(3):
        RHS_Khat.expr += -gam_UU.symb[i][j]*CovDDalpha_LL.symb[j][i]+alpha.symb*Atilde_LL.symb[i][j]*Atilde_UU.symb[i][j]

RHS_Khat.expr += AdvDKhat.symb
RHS_Khat.expr += 1/3*alpha.symb*((Khat.symb+2*theta.symb)**2)+ 1/3*alpha.symb*kappa1.symb*(1-kappa2.symb)*theta.symb
RHS_Khat.expr += KOsigma*dKODFullKhat.symb
```

```python
RHSObjects += kappa1.symb2expr()
RHSObjects += kappa2.symb2expr()
RHSObjects += RHS_Khat.setisymb()
```

```python
RHS_Atilde_LL = stvarrank2('Atilde_LL', sym = 'sym01', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")

for i in range(3):
    for j in range(3):
        RHS_Atilde_LL.expr[i][j] += chi.symb*(-CovDDalphaTF_LL.symb[i][j]+alpha.symb*RTF_LL.symb[i][j])+alpha.symb*(Khat.symb+2*theta.symb)*Atilde_LL.symb[i][j]
        RHS_Atilde_LL.expr[i][j] += AdvDAtilde_LL.symb[i][j]
        for k in range(3):
            RHS_Atilde_LL.expr[i][j] += -2*alpha.symb*Atilde_LL.symb[i][k]*Atilde_UL.symb[k][j]
            RHS_Atilde_LL.expr[i][j] += Atilde_LL.symb[k][i]*dDbeta_UL.symb[k][j]+Atilde_LL.symb[k][j]*dDbeta_UL.symb[k][i]-2/3*Atilde_LL.symb[i][j]*dDbeta_UL.symb[k][k]
            
for i in range(3):
    for j in range(3):
        RHS_Atilde_LL.expr[i][j] += KOsigma*dKODFullAtilde_LL.symb[i][j]
```

```python
RHSObjects += RHS_Atilde_LL.setisymb()
```

```python
RHS_theta = stvar('theta', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")
```

```python
RHS_theta.expr = 1/2*alpha.symb*(Rsclr.symb + 2/3*(Khat.symb+2*theta.symb)**2 - 2*kappa1.symb*(2+kappa2.symb)*theta.symb)
RHS_theta.expr += AdvDtheta.symb
for i in range(3):
    for j in range(3):
        RHS_theta.expr += -1/2*alpha.symb*Atilde_LL.symb[i][j]*Atilde_UU.symb[i][j]

RHS_theta.expr += KOsigma*dKODFulltheta.symb
```

```python
RHSObjects += RHS_theta.setisymb()
```

```python
RHS_Gamtilde_U = stvarrank1('Gamtilde_U', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")

for i in range(3):
    RHS_Gamtilde_U.expr[i] += AdvDGamtilde_U.symb[i]
    for j in range(3):
        RHS_Gamtilde_U.expr[i] += -2*Atilde_UU.symb[i][j]*dDalpha_L.symb[j]+2*alpha.symb*(-1/3*gamtilde_UU.symb[i][j]*(2*dDKhat_L.symb[j]+dDtheta_L.symb[j])-3/2*Atilde_UU.symb[i][j]*dDchi_L.symb[j]/chi.symb)
        RHS_Gamtilde_U.expr[i] += -GamtildeD_U.symb[j]*dDbeta_UL.symb[i][j]+2/3*GamtildeD_U.symb[i]*dDbeta_UL.symb[j][j]
        for k in range(3):
            RHS_Gamtilde_U.expr[i] += 2*alpha.symb*Gamtilde_ULL.symb[i][j][k]*Atilde_UU.symb[j][k]
            RHS_Gamtilde_U.expr[i] += 1/3*gamtilde_UU.symb[i][j]*dDDbeta_ULL.symb[k][j][k]
            RHS_Gamtilde_U.expr[i] += gamtilde_UU.symb[j][k]*dDDbeta_ULL.symb[i][j][k]
            
for i in range(3):
    RHS_Gamtilde_U.expr[i] += -2*alpha.symb*kappa1.symb*(Gamtilde_U.symb[i]-GamtildeD_U.symb[i])
for i in range(3):
    RHS_Gamtilde_U.expr[i] += KOsigma*dKODFullGamtilde_U.symb[i]
```

```python
dDDbeta_ULL.symb[1][0][1]
```

```python
RHSObjects += RHS_Gamtilde_U.setisymb()
```

```python
mul = stvar('mul')
mus = stvar('mus')

mul.expr = 2/alpha.symb
mus.expr = 1/(alpha.symb**(2))
```

```python
RHSObjects += mul.symb2expr()
RHSObjects += mus.symb2expr()
```

```python
RHS_alpha = stvar('alpha', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")

RHS_alpha.expr += -mul.symb*(alpha.symb**2)*Khat.symb+AdvDalpha.symb
RHS_alpha.expr += KOsigma*dKODFullalpha.symb
```

```python
RHSObjects += RHS_alpha.setisymb()
```

```python
eta = stvar('eta')
eta.expr = 2.0

RHS_beta_U = stvarrank1('beta_U', gridvar = True, addtolist = False, vartype = "state_fab", varprefix = "rhs_fab")

for i in range(3):
    RHS_beta_U.expr[i] += mus.symb*(alpha.symb**2)*Gamtilde_U.symb[i]-eta.symb*beta_U.symb[i]+AdvDbeta_U.symb[i]
    
for i in range(3):
    RHS_beta_U.expr[i] += KOsigma*dKODFullbeta_U.symb[i]
```

```python
RHSObjects += eta.symb2expr()
```

```python
RHSObjects += RHS_beta_U.setisymb()
```
