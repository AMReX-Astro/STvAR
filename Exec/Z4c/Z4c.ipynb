```python
var = 'H'
for i in range(0,201):
    ident = list(str(int(i)+10000))
    ident[0] = '0'
    filename = "diag00"+"".join(ident)
    ds = yt.load(filename)
    s = yt.SlicePlot(ds, 'z', ('boxlib', var), origin='native', center = [0,0,0])
    s.set_log(('boxlib', var), False)
    s.set_xlabel('x - Axis')
    s.set_ylabel('y - Axis')
    s.hide_colorbar()
    s.annotate_grids()
    #s.annotate_cell_edges()
    s.zoom(4)
    
    ident2 = list(str(i+10000))
    ident2[0] = '0'
    dst = "imgH"+"".join(ident2)+".png"
    print(filename)
    print(dst)
    s.save(dst)
```
